```python
#create a datafrme dove ogni riga è un vettore di probabilità
import pandas as pd
import numpy as np
import random
import numpy as np
from scipy.special import softmax as sp_softmax
from scipy.spatial.distance import jensenshannon as sp_jensenshannon
import time
from scipy.stats import entropy

def jensenshannon(x, y):
    '''
    Parameters:
    x: input probabilistic array
    y: input probabilistic array
    Returns:
    jensenshannon: Jensen-Shannon divergence along each axis-slice of  the two input arrays

    '''
   # Compute the average of the two arrays
    avg = (x + y) / 2
    # Compute the shannon_entropy of the average array
    shannonentropy_of_avg = entropy(avg)
    #compute the average of the shannon_entropy of the two arrays
    avg_of_shannonentropy = (entropy(x) + entropy(y)) / 2.0
    
    return  (shannonentropy_of_avg-avg_of_shannonentropy)

def jensen_shannon_matrix(x_df, y_df):
    if isinstance(x_df, np.ndarray):
        x_matrix = x_df
    else:
        x_matrix = x_df.values

    if isinstance(y_df, np.ndarray):
        y_matrix = y_df
    else:
        y_matrix = y_df.values

    results = np.zeros((x_matrix.shape[0], y_matrix.shape[0]))
    
    for i in range(x_matrix.shape[0]):
        for j in range(y_matrix.shape[0]):
            results[i, j] = jensenshannon(x_matrix[i], y_matrix[j])
            
    return results

def sp_jensen_shannon_matrix(x_df, y_df):
    if isinstance(x_df, np.ndarray):
        x_matrix = x_df
    else:
        x_matrix = x_df.values

    if isinstance(y_df, np.ndarray):
        y_matrix = y_df
    else:
        y_matrix = y_df.values

    results = np.zeros((x_matrix.shape[0], y_matrix.shape[0]))
    
    for i in range(x_matrix.shape[0]):
        for j in range(y_matrix.shape[0]):
            results[i, j] = sp_jensenshannon(x_matrix[i], y_matrix[j])
            
    return results

def sp_squared_jensen_shannon_matrix(x_df, y_df):
    if isinstance(x_df, np.ndarray):
        x_matrix = x_df
    else:
        x_matrix = x_df.values

    if isinstance(y_df, np.ndarray):
        y_matrix = y_df
    else:
        y_matrix = y_df.values

    results = np.zeros((x_matrix.shape[0], y_matrix.shape[0]))
    
    for i in range(x_matrix.shape[0]):
        for j in range(y_matrix.shape[0]):
            results[i, j] = sp_jensenshannon(x_matrix[i], y_matrix[j])**2
            
    return results
def old_sed(x, y):
    '''
    Parameters:
    x: input probabilistic array 
    y: input probabilistic array 
    Returns:
    sed: symmetric entropy distance between the two input arrays
    '''
    # Compute the average of the two arrays
    avg = (x + y) / 2
    # Compute the complexity of the average array
    complexity_avg = complexity(avg)
    # Compute the product of the complexity of the two arrays
    complexity_product = (complexity(x) * complexity(y)) ** (1 / 2)
    # Compute the symmetric entropy distance
    sed = complexity_avg / complexity_product
    return sed-1

def complexity(x):
    '''
    Parameters:
    x: input probabilistic array 
    Returns:
    complexity: complexity of the input array
    '''
    return np.exp(shannon_entropy(x))

def shannon_entropy(x):
    '''
    Parameters:
    x: input probabilistic array 
    Returns:
    shannon_entropy: shared entropy of the input array
    '''
    y= np.where(x == 0, 1, x)  # Replace zero or negative values with epsilon
    ##shall we do #epsilon = 1e-10  # Small epsilon to avoid zero or negative values y= np.where(x < epsilon, 1, x)  # x = np.where(x <= 0, epsilon, x)  # Replace zero or negative values with epsilon
    if np.any(y <= 0):
        raise ValueError("Invalid values encountered in input array x")
    log_x = np.log(y)
    shannon_entropy = -np.sum(x * log_x)
    return shannon_entropy

def sed(x, y):
    '''
    Parameters:
    x: input probabilistic array 
    y: input probabilistic array 
    Returns:
    sed: symmetric entropy distance along each axis-slice of x  the two input arrays
    '''
    return np.exp(jensenshannon(x,y))-1
def sed_matrix(x_df, y_df):
    if isinstance(x_df, np.ndarray):
        x_matrix = x_df
    else:
        x_matrix = x_df.values

    if isinstance(y_df, np.ndarray):
        y_matrix = y_df
    else:
        y_matrix = y_df.values

    results = np.zeros((x_matrix.shape[0], y_matrix.shape[0]))
    
    for i in range(x_matrix.shape[0]):
        for j in range(y_matrix.shape[0]):
            results[i, j] = sed(x_matrix[i], y_matrix[j])
            
    return results

def old_sed_matrix(x_df, y_df):
    if isinstance(x_df, np.ndarray):
        x_matrix = x_df
    else:
        x_matrix = x_df.values

    if isinstance(y_df, np.ndarray):
        y_matrix = y_df
    else:
        y_matrix = y_df.values

    results = np.zeros((x_matrix.shape[0], y_matrix.shape[0]))
    
    for i in range(x_matrix.shape[0]):
        for j in range(y_matrix.shape[0]):
            results[i, j] = old_sed(x_matrix[i], y_matrix[j])
            
    return results
x_df=pd.DataFrame(np.random.rand(2,10))
y_df=pd.DataFrame(np.random.rand(2,10))

#apply softmax at each row of x_df
x_df=x_df.apply(sp_softmax,axis=1)
#apply softmax at each row of y_df
y_df=y_df.apply(sp_softmax,axis=1)
print(x_df.shape)

#print("x_df=\n",x_df, "\n shape=",x_df.shape)
#print("y_df=\n",y_df,"\n shape=",y_df.shape)
#time to cumpute the jensen shannon distance
print("sp_jensenshannon")
startms = time.time()
scipy_js=sp_jensen_shannon_matrix(x_df, y_df)
endms = time.time()
print(f"-->sp_js matrix={scipy_js}")
#print shape of the matrix
print(f"shape={scipy_js.shape}")
print(f"sp_jensenshannon time={endms-startms}")

print("scipy squared jensenshannon")
startms = time.time()
scipy_js_sq=sp_squared_jensen_shannon_matrix(x_df, y_df)
endms = time.time()
print(f"-->scipy_js_sq matrix={scipy_js_sq}")
print(f"shape={scipy_js_sq.shape}")
print(f"scipy squared jensenshannon time={endms-startms}")

print("ours")
startms = time.time()
our_js=jensen_shannon_matrix(x_df, y_df )
endms = time.time()
print(f"-->our_js matrix={our_js}")
print(f"shape={our_js.shape}")

print(f"our_js time={endms-startms}")

scipy_js2=scipy_js**2
print ("-->scipy_js**2=",scipy_js2)
print ("shape=",(scipy_js2).shape)  

print("**")


print("diff")


#print(f"our_js-scipy_js: {(our_js-scipy_js).max()}")
print(f"our_js-scipy_js_sq: {(our_js-scipy_js_sq).max()}")

print(f"our_js-scipy_js**2 {(our_js-(scipy_js2)).max()}")
print(f"scipy_js**2 -scipy_js_sq{((scipy_js2 )-scipy_js_sq).max()}")

# print("sp_jensenshannon")
# print(jensen_shannon_matrix(x_df, y_df )**2)
# print("ours")
# print(jensen_shannon_matrix_test(x_df, y_df ))
# print("diff")
# print(f"{jensen_shannon_matrix_test(x_df, y_df)-(jensen_shannon_matrix(x_df, y_df)**2)}")

#old sed
print("old sed")
startms = time.time()
old_sed=old_sed_matrix(x_df, y_df)
endms = time.time()
print(f"-->old_sed matrix={old_sed}")
print(f"shape={old_sed.shape}")
print(f"old_sed time={endms-startms}")

print("ours")
startms = time.time()
our_sed=sed_matrix(x_df, y_df)
endms = time.time()
print(f"-->our_sed matrix={our_sed}")
print(f"shape={our_sed.shape}")
print(f"our_sed time={endms-startms}")

print("diff")
print(f"our_sed-old_sed: {(our_sed-old_sed).max()}")



```
